Imports

```python
import pandas as pd
import json
# import yfinance as yf
```

Load csv file

```python
csv = pd.read_csv('securitiesinfo.csv', skiprows=2)
```

Open the json file and append the data

```python
symbol_storage = open('indicator_international_symbols.json', 'r')
data = json.load(symbol_storage)
symbol_storage.close()

for index, row in csv.iterrows():
    try:
        indicator = str(row['Security No'])
        # name_he = row['Name (Hebrew)']
        name = row['Name (English)']
        international_symbol = row['Symbol (English)']

        # First remove the .TA suffix
        if '.TA' in international_symbol:
            international_symbol = international_symbol.replace('.TA', '')
            
        # Replace any dots with hyphens and append .TA
        international_symbol = international_symbol.replace('.','-')
        international_symbol = international_symbol + '.TA'
        
        ISIN = csv.iloc[index, 1] # ISIN is in the second column

        data[indicator] = { "name": name,
                            "symbol": international_symbol, 
                            "ISIN": ISIN}
    except Exception as e:
        print(f"Error processing row {index}: {e}")

# Load archive
f_archive = open('international_symbols_archive.json', 'r')
archive = json.load(f_archive)
f_archive.close()

# Merge archive into data
data.update(archive)
# No duplicate keys will exist; existing keys will be overwritten.
```

Check existence of 1183441

```python
print(data.keys())
print('1209790' in data.keys())
print('1183441' in data.keys())
```

Save json file

```python
# Save data in indicator_international_symbols.json
with open('indicator_international_symbols.json', 'w') as f:
    f.write(json.dumps(data, indent=4))
```
